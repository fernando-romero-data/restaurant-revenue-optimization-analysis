## 7.1 Off-Peak Revenue Activation Experiment

- Objective: Increase revenue during underperforming mid-afternoon hours (13–16).
- Prior analysis showed:
  - low order volume
  - lower AOV
  - minimal operational pressure
- This time window represents the largest opportunity for incremental growth without increasing fixed costs.

### Hypothesis

- If customers are encouraged to treat mid-afternoon as a “meal occasion” rather than a snack window, both order volume and AOV will increase.

### Experiment Design

- Introduce a time-based promotion or positioning between 13:00–16:00:
  - example: “Afternoon Meal Combo”
  - focus on perceived value rather than heavy discounting
- Emphasize complete meal behavior:
  - entrée + drink
  - entrée + side

- Ensure operational simplicity:
  - use items already prepared or easy to assemble
  - avoid adding complexity during peak prep cycles

### Success Metrics

- Increase in total orders during 13–16 window
- Increase in average order value (AOV) during 13–16
- Increase in total revenue for off-peak hours
- No negative impact on peak-hour operations

### Measurement Approach

- Compare pre- and post-experiment performance:
  - orders/hour
  - AOV
  - revenue/hour
- Track results over a defined period (e.g., 2–4 weeks)
- Use historical baseline from Workbook 3 and 5 for comparison

### Expected Impact

- Improved utilization of underperforming hours
- Increased total daily revenue without additional fixed costs
- More balanced demand distribution across the day

## 7.2 AOV Growth Experiment (Single-Customer Focus)

- Objective: Increase Average Order Value (AOV) across the majority of transactions.
- Prior analysis showed:
  - most orders are single-guest
  - AOV increases significantly with guest count
  - revenue is volume-driven rather than concentrated

- This indicates that small increases in AOV across many orders can drive meaningful revenue growth.

### Hypothesis

- If low-friction add-ons are consistently prompted during ordering, single-customer AOV will increase without reducing order volume.

### Experiment Design

- Introduce standardized upsell prompts at point-of-sale:
  - “Would you like to add a drink?”
  - “Add fries for $X?”
- Focus on:
  - high-margin items
  - quick-add items that do not slow service

- Optional:
  - bundle light combinations (entrée + drink)
  - avoid heavy discounting

### Success Metrics

- Increase in AOV for single-guest orders
- Increase in attach rate (add-on items per order)
- Increase in total revenue per hour

### Measurement Approach

- Compare:
  - AOV before vs after implementation
  - % of orders with add-ons
- Segment specifically:
  - single-guest transactions (primary target)

### Expected Impact

- Incremental revenue growth across a large portion of transactions
- Higher profitability due to add-on margins
- Minimal operational disruption

## 7.3 Peak Hour Efficiency Experiment

- Objective: Maximize revenue capture during peak demand periods (11–12 and 17–18).
- Prior analysis showed:
  - ~76% of revenue occurs during peak hours
  - demand is concentrated in short time windows
  - operational constraints during these periods can limit throughput

- This indicates that improving efficiency during peak hours can directly increase revenue without increasing demand.

### Hypothesis

- If order throughput is improved during peak periods, more orders can be fulfilled, increasing total revenue without requiring additional customers.

### Experiment Design

- Reduce friction in order processing during peak hours:
  - simplify menu presentation (highlight fastest-moving items)
  - streamline order-taking process
  - pre-prep high-demand ingredients before peak

- Implement operational adjustments:
  - assign clear roles (cashier, prep, cook)
  - prioritize speed over customization during peak
  - reduce decision complexity for customers

- Optional:
  - introduce “fast lane” items (limited menu during rush)
  - pre-build common combinations

### Success Metrics

- Increase in orders per hour during peak windows
- Reduction in average service time
- Increase in revenue per peak hour
- No decrease in customer satisfaction

### Measurement Approach

- Compare peak-hour performance before vs after:
  - orders/hour
  - revenue/hour
- Monitor service time if available
- Validate that improvements are consistent across multiple days

### Expected Impact

- Higher revenue capture during peak periods
- Reduced bottlenecks and operational pressure
- Improved customer experience through faster service

## 7.4 Demand Smoothing Experiment (Shift Peak Load)

- Objective: Reduce peak-hour congestion by shifting a portion of demand slightly earlier or later.
- Prior analysis showed:
  - heavy concentration of demand in 11–12 and 17–18 windows
  - operational pressure during these periods
  - underutilized capacity just outside peak windows

- This creates an opportunity to redistribute demand without losing revenue.

### Hypothesis

- If customers are incentivized to order slightly before or after peak hours, demand will spread more evenly, reducing bottlenecks and increasing total orders fulfilled.

### Experiment Design

- Introduce **time-adjacent incentives**:
  - “Early Lunch Special (10:30–11:00)”
  - “Early Dinner Offer (16:30–17:00)”
- Keep incentives light (value framing, small add-on) to avoid margin erosion.
- Emphasize convenience:
  - shorter wait times
  - faster service outside peak

- Operational alignment:
  - ensure readiness slightly before peak
  - maintain staffing coverage in adjacent windows

### Success Metrics

- Increase in orders during pre-peak and post-peak windows
- Reduction in order concentration during peak hours
- Increase in total daily orders fulfilled
- Stable or improved AOV

### Measurement Approach

- Compare distribution of orders by hour before vs after experiment
- Track:
  - orders/hour shift into adjacent windows
  - peak-hour load reduction
- Evaluate over a 2–4 week period

### Expected Impact

- Smoother demand curve across the day
- Reduced operational bottlenecks during peak hours
- Increased total throughput and revenue without adding capacity

## 7.6 SQL Segment Analysis (Experiment Setup)

- Segment orders into meaningful time-based groups (lunch, dinner, off-peak).
- Compare performance across segments using SQL.
- Establish a foundation for testing strategies (e.g., off-peak promotions, AOV bundles).

```python
import pandas as pd
import duckdb

# ==============================
# Step 1: Load cleaned dataset
# ==============================

orders_clean = pd.read_csv("../data/cleaned/orders_clean.csv")

# restore datetime (CSV limitation)
orders_clean["order_datetime"] = pd.to_datetime(orders_clean["order_datetime"])

# ==============================
# Step 2: DuckDB connection
# ==============================

con = duckdb.connect()
con.register("orders_clean", orders_clean)

# ==============================
# Step 3: SQL Segmentation
# ==============================

con.execute("""
SELECT
    CASE
        WHEN EXTRACT(HOUR FROM order_datetime) BETWEEN 11 AND 13 THEN 'lunch'
        WHEN EXTRACT(HOUR FROM order_datetime) BETWEEN 17 AND 19 THEN 'dinner'
        ELSE 'off_peak'
    END AS time_segment,
    
    COUNT(*) AS total_orders,
    SUM(order_total) AS total_revenue,
    ROUND(AVG(order_total), 2) AS avg_order_value

FROM orders_clean

GROUP BY time_segment
ORDER BY total_revenue DESC
""").fetchdf()
```

```python
con.execute("""
SELECT
    CASE
        WHEN EXTRACT(HOUR FROM order_datetime) BETWEEN 11 AND 13 THEN 'Lunch'
        WHEN EXTRACT(HOUR FROM order_datetime) BETWEEN 17 AND 19 THEN 'Dinner'
        ELSE 'Off-Peak'
    END AS time_segment,

    CASE
        WHEN order_total < 15 THEN 'Low Value (<$15)'
        WHEN order_total BETWEEN 15 AND 30 THEN 'Mid Value ($15–$30)'
        ELSE 'High Value ($30+)'
    END AS value_segment,

    COUNT(*) AS total_orders,
    SUM(order_total) AS total_revenue,
    ROUND(AVG(order_total), 2) AS avg_order_value

FROM orders_clean

GROUP BY time_segment, value_segment

ORDER BY 
    time_segment,
    value_segment
""").fetchdf()
```
