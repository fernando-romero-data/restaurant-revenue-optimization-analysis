# 10.0 Owner Strategy Playbook Overview

- This playbook shows how to increase revenue using your existing customers
- No marketing spend required
- Focus on:
  - Increasing order size
  - Improving slow hours
  - Capturing more during busy hours

## Bottom Line

- You already have demand
- The opportunity is getting more value from each order

# 10.1 Revenue Reality — Where the Money Comes From

## What’s Happening

- Most revenue comes from larger orders (30+)
- Smaller orders happen often but don’t make much money

- Peak hours (lunch and dinner):
  - Drive most of the revenue

- Off-peak hours:
  - Have customers, but low spend

## What This Means

- You don’t need more customers
- You need:
  - Bigger orders
  - Better use of slow hours

## Key Insight

- This is a revenue capture problem, not a demand problem

# 10.2 AOV Playbook — Increase Every Order

## Goal

- Get each customer to spend a little more

## How to Do It

- Add simple items to every order:
  - Cheese dip
  - Chips
  - Drinks

- Use small add-ons (3 to 5 range)
  - Easy for customers to say yes

## What to Say (POS or Staff)

- Add cheese dip?
- Add a drink?

## Why This Works

- Customers are already close to spending more
- Small adds move orders into higher price ranges

## Example

- Small order becomes a full meal with one add-on
- Mid-size order becomes a high-value order with one extra item

## Result

- More revenue from the same number of customers

# 10.3 Menu Engineering — What to Push and What to Fix

## What’s Driving Revenue

- A small number of items make most of your money
- Top items:
  - Nelson Special
  - Burrito California
  - Cheese dip

## What to Do

- Put these items at the top of your menu
- Label them:
  - Most Popular
  - Customer Favorite
- Make them easy to order

## Add-On Strategy (Big Win)

- Cheese dip is your best upsell item
- Add it to every order opportunity

- Train staff or POS to ask:
  - Add cheese dip?
  - Add chips and salsa?
  - Add a drink?

## Drinks Strategy

- Drinks sell a lot but are priced low
- Small price increase = big revenue impact

## Fix Menu Issues

- Clean up duplicate items (same item, different names)
- Keep menu simple and focused

## Goal

- Push customers toward high-performing items
- Increase add-ons on every order
- Remove confusion in the menu

## Result

- Higher average order value
- More consistent revenue from top items

# 10.4 Peak Hours — Capture More Revenue

## What’s Happening

- Lunch and dinner are your busiest times
- Most of your revenue is made during these hours

## Problem

- You’re limited by how fast you can make food
- Slow service = lost orders

## What to Do

- Speed up the kitchen:
  - Prep ingredients before rush
  - Focus on fast, repeatable items

- Simplify orders during peak:
  - Push standard meals
  - Avoid too many custom requests

- Make ordering faster:
  - Clear menu
  - Easy combo options

## Staffing

- Make sure you are fully staffed during peak
- Focus on:
  - Kitchen speed
  - Order flow

## Goal

- Serve more customers per hour
- Reduce wait times
- Avoid losing orders during rush

## Result

- More orders completed
- Higher revenue during busiest hours

# 10.5 Off-Peak Hours — Turn Slow Time Into Revenue

## What’s Happening

- Mid-afternoon (13 to 16) is slow
- Customers are ordering small, low-value items

## Opportunity

- Kitchen is not busy
- Staff is available
- Any increase in order size = extra profit

## What to Do

- Push simple “meal” orders:
  - Entree plus drink
  - Entree plus side

- Use light incentives:
  - Free add-on after minimum spend
  - Small discount to reach next price level

- Highlight add-ons:
  - Cheese dip
  - Chips
  - Drinks

## Goal

- Turn small orders into full meals
- Increase spend per customer

## Result

- More revenue during slow hours
- Better use of staff and kitchen time

# 10.6 Pricing Strategy — Get Customers to Spend More

## What’s Happening

- Customers tend to stop spending at certain points:
  - Around 12 to 14
  - Around 25 to 29

## Opportunity

- Customers close to these levels are easiest to increase
- Small changes can push them to spend more

## What to Do

- Use add-ons priced in the 3 to 5 range
  - Easy yes for customers

- Set up pricing to encourage the next step:
  - Make it easy to go from 25 to 30+
  - Make it easy to go from 12 to 15+

- Use simple prompts:
  - You’re close to a full meal, add one more item?
  - Add cheese dip or a drink?

## Menu Setup

- Highlight items that help complete the order:
  - Cheese dip
  - Chips
  - Drinks

- Position full meals slightly above the next level

## Goal

- Move more orders into higher price ranges
- Increase average order value without raising base prices

## Result

- More high-value orders
- More revenue from the same customers

  - Cheese dip
  - Chips
  - Drinks

- More orders become full meals instead of single items

## Revenue Impact

## Bottom Line

- The business already has demand

- Simple changes can unlock significant revenue

# 10.9 Menu Design — How to Structure the Menu for More Revenue

## Core Idea

- Customers should order meals, not single items
- The menu should guide them to spend more without thinking about it

## New Menu Structure

### Meals (Top of Menu)

- Burrito Meal (burrito + drink + cheese dip)
- Taco Meal (3 tacos + drink + cheese dip)
- Quesadilla Meal (quesadilla + drink + side)

- These should be the first options customers see

## Favorites Section

- Nelson Special
- Burrito California
- Taco Plate

- Label as:
  - Most Popular
  - Customer Favorite

## Build Your Own (Lower on Menu)

- Burrito
- Tacos
- Quesadilla

- Always pair with add-on prompts:
  - Add cheese dip
  - Add a drink

## Add-Ons (Highly Visible)

- Cheese dip (primary upsell)
- Chips
- Drinks

- These should not be hidden
- They should appear during ordering

## Pricing Strategy

- Keep add-ons in the 3 to 5 range
- Position meals just above key thresholds:
  - Around 15
  - Around 30

## Off-Peak Adjustment

- Highlight simple meal options during slow hours
- Encourage full meals instead of small orders

## Goal

- Turn single-item orders into full meals
- Increase average order value across all customers

## Result

- More add-ons per order
- More high-value orders
- Higher revenue without increasing traffic
