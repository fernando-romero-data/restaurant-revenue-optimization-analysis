# 9.0 Executive Report — Revenue Optimization System

## 9.1 Objective

- Translate analytical findings into actionable business strategies
- Identify where revenue is being generated vs. underutilized
- Focus on increasing revenue without increasing traffic
- Align operational execution with observed demand patterns

---

## 9.2 Data Scope & Constraints

- Analysis is based on order-level dataset (`orders_clean`)
- No direct item-level linkage is available
- Menu strategies are derived through logical inference, not direct SKU analysis
- All recommendations are grounded in observed patterns:
  - Order value segmentation
  - Time-based demand behavior
  - Revenue concentration

---

## 9.3 Strategic Framing

- Revenue growth will not come from more customers
- Revenue growth will come from:

  - Increasing Average Order Value (AOV)
  - Shifting low-value orders into higher tiers
  - Improving demand utilization across time

---

## 9.4 Core Business Insight

- High-value orders (30+ dollars) drive the majority of revenue
- Mid-value orders (15–30 dollars) represent the largest opportunity for conversion
- Low-value orders (under 15 dollars) dominate off-peak periods

**Implication:**

- The business has sufficient demand  
- The primary opportunity is **order value expansion**, not traffic acquisition

## 9.5 AOV Expansion Strategy — Add-On Driven Growth

- High-value orders (30+ dollars) drive the majority of revenue
- Mid-value orders (15–30 dollars) represent the largest opportunity for conversion
- Small increases in order size can shift a significant portion of orders into higher revenue tiers

**Strategy Focus:**

- Increase Average Order Value (AOV) without increasing traffic
- Use low-friction add-ons to move orders across value thresholds

---

## 9.5.1 Add-On Economics (Why This Works)

- Customers are already purchasing within close range of higher value tiers
- Incremental spend of 3–6 dollars can shift order classification
- Add-ons require minimal decision effort compared to full item upgrades

**Key Principle:**

- Do not force large spending decisions  
- Instead, introduce small, natural purchase extensions

---

## 9.5.2 Cheese Dip as a Primary AOV Lever

- Cheese dip is a low-cost, high-margin add-on
- Naturally complements core menu items:
  - Tacos
  - Burritos
  - Quesadillas
  - Combination plates

- Requires minimal additional preparation or operational complexity

**Strategic Role:**

- Acts as a “bridge item” to increase total order value
- Enables consistent AOV lift across all order types

---

## 9.5.3 Checkout Prompt Strategy

- Introduce a default add-on prompt during ordering:

  Example:
  - “Add cheese dip for 3.99?”

- Position as a complement to the meal, not an upsell

**Expected Impact:**

- Increases attachment rate across both mid-value and low-value orders
- Generates incremental revenue with minimal friction

---

## 9.5.4 Threshold-Based Upsell Mechanics

- Use add-ons to push orders across value tiers

**Example Scenarios:**

- Order at 27 dollars → add cheese dip → moves to 31 dollars → becomes high-value order  
- Order at 13 dollars → add cheese dip → moves to 17 dollars → becomes mid-value order  

**Outcome:**

- Systematically increases the proportion of higher-value transactions

---

## 9.5.5 Bundle Integration (Menu-Consistent)

- Integrate cheese dip into natural meal groupings

**Example:**

- Burrito + drink + cheese dip  
- Taco plate + cheese dip add-on  

**Design Principles:**

- Keep bundles aligned with existing ordering behavior
- Avoid introducing unfamiliar combinations
- Use add-ons to enhance, not replace, core items

---

## 9.5.6 Business Impact

- Increase in Average Order Value (AOV)
- Higher share of orders in 30+ dollar category
- Revenue growth without additional customer acquisition
- Minimal operational disruption

---

## 9.5.7 System-Level Insight

- AOV growth is most effectively driven by:

  - High-margin add-ons
  - Threshold-based nudging
  - Low-friction purchase decisions

**Conclusion:**

- Small, repeatable add-on strategies create scalable revenue impact

## 9.6 Off-Peak Revenue Strategy — Demand Shaping

- Off-peak periods are dominated by low-value orders (under 15 dollars)
- Demand exists but is under-monetized
- Capacity is available during these hours

**Strategy Focus:**

- Increase revenue during off-peak without hurting peak performance
- Shift low-value orders into mid-value tiers
- Utilize existing operational capacity

---

## 9.6.1 Off-Peak Behavior Insight

- Customers during off-peak hours tend to:
  - Order fewer items
  - Spend less per transaction
  - Make more price-sensitive decisions

**Implication:**

- Off-peak is not a demand problem  
- It is a **value capture problem**

---

## 9.6.2 Time-Based Promotion Strategy

- Introduce promotions only during low-demand windows

**Example Window:**

- Mid-afternoon (e.g., 2PM–5PM)

**Tactics:**

- Minimum spend incentives:
  - “Free cheese dip with orders over 20 dollars”
- Light discount thresholds:
  - “3 dollars off orders over 25 dollars”

**Design Principle:**

- Promotions must increase order value, not reduce it

---

## 9.6.3 AOV-Driven Incentive Design

- Structure offers to push orders into higher tiers

**Example Mechanics:**

- Customer at 14 dollars → incentivized to reach 20+  
- Customer at 22 dollars → nudged toward 25+  

**Outcome:**

- Converts low-value orders into mid-value orders
- Increases overall revenue per transaction

---

## 9.6.4 Menu Positioning During Off-Peak

- Highlight shareable and add-on friendly items:
  - Combination plates
  - Add-ons (cheese dip, sides, drinks)

- Reinforce “value” perception rather than discount perception

**Goal:**

- Encourage customers to build larger orders naturally

---

## 9.6.5 Operational Advantage

- Off-peak hours have:
  - Lower kitchen load
  - Faster prep times
  - More available staff attention

**Opportunity:**

- Capture additional revenue without adding labor cost
- Improve overall daily revenue distribution

---

## 9.6.6 Guardrails

- Do not apply promotions during peak hours
- Avoid deep discounts that reduce margins
- Ensure offers maintain profitability thresholds

---

## 9.6.7 Business Impact

- Increased revenue during previously underutilized hours
- Higher Average Order Value (AOV) during off-peak
- Improved efficiency of existing operations
- Better demand distribution across the day

---

## 9.6.8 System-Level Insight

- Off-peak revenue growth is driven by:

  - Time-targeted incentives
  - Threshold-based pricing
  - Strategic use of high-margin add-ons

**Conclusion:**

- Off-peak is the largest untapped revenue opportunity when properly structured

## 9.7 Peak Hour Optimization — Throughput & Revenue Capture

- Peak hours generate the highest concentration of orders and revenue
- High-value orders (30+ dollars) are most common during these periods
- Demand is already strong and consistent

**Strategy Focus:**

- Maximize revenue capture during peak demand
- Reduce operational bottlenecks
- Increase total orders fulfilled per hour

---

## 9.7.1 Peak Hour Constraint Insight

- Peak periods are not limited by demand
- They are limited by operational throughput:

  - Kitchen capacity
  - Order processing speed
  - Staff coordination

**Implication:**

- Revenue is being lost due to inability to fulfill additional orders
- This is a **capacity constraint problem**, not a marketing problem

---

## 9.7.2 Throughput Optimization

- Focus on increasing the number of orders completed per hour

**Tactics:**

- Streamline high-frequency items (faster prep workflows)
- Pre-prep ingredients before peak windows
- Reduce complexity in peak-hour order flow

**Outcome:**

- More orders fulfilled without increasing demand
- Higher revenue during the busiest hours

---

## 9.7.3 Menu Execution Strategy

- Identify and prioritize operationally efficient items during peak

**Approach:**

- Highlight items that:
  - Have faster preparation time
  - Require fewer steps
  - Maintain strong margins

**Example:**

- Promote combination plates or standard builds over highly customized orders

**Goal:**

- Maintain speed without sacrificing order value

---

## 9.7.4 Order Flow Optimization

- Reduce friction in how orders are received and processed

**Tactics:**

- Simplify ordering interface (POS or online)
- Limit excessive customization during peak windows
- Improve handoff between front-of-house and kitchen

**Outcome:**

- Faster order turnaround
- Reduced backlog during high-demand periods

---

## 9.7.5 Staffing Alignment

- Align labor with demand spikes

**Focus Areas:**

- Ensure adequate kitchen staffing during peak hours
- Optimize role allocation:
  - Order intake vs food preparation
- Minimize idle time outside peak while maximizing coverage during peak

---

## 9.7.6 Guardrails

- Do not introduce promotions during peak hours
- Avoid adding complexity to the menu during high-demand periods
- Maintain consistency and speed as priority

---

## 9.7.7 Business Impact

- Increased total orders fulfilled during peak hours
- Higher revenue without additional customer acquisition
- Reduced lost sales due to operational constraints
- Improved customer experience (faster service times)

---

## 9.7.8 System-Level Insight

- Peak revenue growth is driven by:

  - Throughput efficiency
  - Operational simplicity
  - Staff alignment with demand

**Conclusion:**

- Optimizing operations during peak hours unlocks immediate revenue without changing demand levels

## 9.8 Pricing & Menu Structuring — Threshold-Based Optimization

- Order value segmentation shows clear spending tiers:
  - Under 15 dollars (low-value)
  - 15–30 dollars (mid-value)
  - 30+ dollars (high-value)

- Customer behavior naturally clusters around these thresholds

**Strategy Focus:**

- Use pricing and menu structure to guide customers into higher value tiers
- Increase perceived value without relying on heavy discounting

---

## 9.8.1 Pricing Threshold Insight

- Small price differences can significantly influence customer decisions
- Customers are more likely to increase spend when close to a threshold

**Example Behavior:**

- Customer at 27 dollars is more likely to add 1 item than a customer at 18 dollars

**Implication:**

- Strategic pricing should focus on crossing thresholds, not maximizing single item price

---

## 9.8.2 Price Laddering

- Tighten the gap between mid-value and high-value tiers

**Approach:**

- Position common order builds just above 30 dollars
- Ensure add-ons can easily move orders from:
  - 25 → 30+
  - 12 → 15+

**Goal:**

- Make higher-value orders feel like the natural next step

---

## 9.8.3 Menu Framing & Positioning

- Highlight items that support AOV growth:

  - Most Popular
  - Best Value
  - Customer Favorite

- Guide customers toward higher-value combinations through placement and labeling

**Outcome:**

- Reduces decision friction
- Increases likelihood of higher spend

---

## 9.8.4 Bundle Structuring (Flexible, Not Rigid)

- Avoid rigid combo-only systems
- Allow customers to build their own higher-value orders

**Example Structure:**

- Core item (burrito or taco plate)  
- + Add-on (cheese dip, side, drink)  
- Naturally reaches higher price tier

**Design Principle:**

- Enhance existing behavior rather than forcing new patterns

---

## 9.8.5 Psychological Pricing

- Use price points that encourage incremental spend

**Examples:**

- 29 instead of 30 to trigger add-on behavior  
- Add-ons priced in low-friction range (3–5 dollars)

**Outcome:**

- Encourages completion of higher-value orders
- Increases attachment rate of add-ons

---

## 9.8.6 Menu-Driven Conversion Strategy

- Orders tend to cluster just below key thresholds:
  - ~12–14 dollars (low → mid conversion zone)
  - ~25–29 dollars (mid → high conversion zone)

- Menu should be structured to push customers across these thresholds

---

### 9.8.6.1 Low → Mid Conversion (12 → 15+ dollars)

**Customer Behavior:**

- Typically ordering a single core item
- Low attachment of sides or add-ons

**Menu Strategy:**

- Introduce low-friction add-ons in the 3–5 dollar range:

  - Cheese dip
  - Chips and salsa upgrade
  - Drink pairing

**Execution:**

- Prompt at checkout:
  - “Add chips and cheese dip for 3.99?”

**Outcome:**

- Moves orders into mid-value tier with minimal resistance
- Increases total revenue per transaction

---

### 9.8.6.2 Mid → High Conversion (25 → 30+ dollars)

**Customer Behavior:**

- Already ordering multiple items
- Close to high-value threshold but not crossing it

**Menu Strategy:**

- Introduce “completion” items:

  - Cheese dip (primary lever)
  - Side add-ons (rice, beans)
  - Drink pairing

**Execution:**

- Prompt near threshold:
  - “You’re 3 dollars away from a full meal — add cheese dip?”

**Outcome:**

- Converts mid-value orders into high-value orders
- Increases revenue concentration in top tier

---

### 9.8.6.3 Soft Bundle Design

- Build natural combinations instead of rigid combos

**Examples:**

- Burrito + drink + cheese dip  
- Taco plate + chips + cheese dip  

**Design Principles:**

- Match existing ordering behavior
- Avoid unfamiliar combinations
- Maintain flexibility for customization

---

## 9.8.7 Guardrails

- Avoid over-discounting that reduces margins
- Maintain consistency in pricing structure
- Keep menu simple and easy to navigate

---

## 9.8.8 Business Impact

- Increased conversion of mid-value to high-value orders
- Higher Average Order Value (AOV)
- Improved revenue without increasing traffic
- Stronger alignment between pricing and customer behavior

---

## 9.8.9 System-Level Insight

- Pricing optimization is most effective when combined with:

  - Add-on strategies (9.5)
  - Off-peak incentives (9.6)
  - Peak operational efficiency (9.7)

**Conclusion:**

- Pricing and menu structure act as the control system that guides customer spending behavior

## 9.9 Menu Architecture — Upstream Revenue Design

- Previous sections focused on optimizing existing behavior
- This section focuses on designing the menu to create desired behavior

**Core Shift:**

- From reacting to customer decisions  
- To structuring the menu so customers naturally spend more

---

## 9.9.1 Upstream vs Downstream Strategy

- Downstream optimization:
  - Add-on prompts
  - Promotions
  - Pricing adjustments

- Upstream design:
  - Default order structures
  - Menu layout
  - Built-in value pathways

**Insight:**

- The most scalable revenue gains come from upstream design, not downstream correction

---

## 9.9.2 Designing for Target Order Tiers

- Order data shows clear revenue thresholds:
  - 15 dollars (low → mid transition)
  - 30 dollars (mid → high transition)

**Menu Objective:**

- Make 15+ dollars and 30+ dollars the natural outcome of most orders

---

## 9.9.3 Default Meal Structuring

- Shift from item-based ordering to meal-based ordering

**Traditional Flow:**

- Customer selects a single item → may add extras

**Optimized Flow:**

- Customer starts with a “meal” structure:

  - Core item (burrito, taco plate)  
  - + Drink  
  - + Add-on (cheese dip or side)

**Outcome:**

- Higher starting order value
- Reduced reliance on upsell prompts

---

## 9.9.4 Embedded Add-On Strategy

- Integrate high-margin add-ons into the default experience

**Example:**

- Cheese dip positioned as:
  - Standard pairing option
  - Suggested inclusion in meal builds

**Design Principle:**

- Add-ons should feel expected, not optional

**Impact:**

- Increases attachment rate automatically
- Drives consistent AOV growth

---

## 9.9.5 Menu Layout & Decision Flow

- Organize menu to guide customer decisions

**Approach:**

- Group items into:
  - Single items
  - Meal builds
  - Best value combinations

- Highlight pathways that lead to higher-value orders

**Goal:**

- Reduce decision fatigue
- Direct customers toward optimal choices

---

## 9.9.6 Pricing Alignment with Structure

- Ensure pricing reinforces menu design

**Tactics:**

- Position full meal builds just above 30 dollars
- Keep add-ons in low-friction range (3–5 dollars)
- Align price gaps with threshold conversion strategy

**Outcome:**

- Pricing and structure work together to drive behavior

---

## 9.9.7 Operational Alignment

- Menu design must align with kitchen execution

**Considerations:**

- Avoid adding complexity during peak hours
- Use repeatable, easy-to-execute components
- Ensure add-ons (like cheese dip) are operationally efficient

---

## 9.9.8 System-Level Insight

- Sustainable revenue growth comes from:

  - Designing the menu for AOV expansion
  - Embedding add-ons into default behavior
  - Aligning pricing, operations, and customer flow

**Conclusion:**

- The menu is not just a list of items  
- It is the primary system that controls customer spending behavior

---

## 9.9.9 Final Strategic Takeaway

- Downstream tactics increase revenue incrementally  
- Upstream design increases revenue systematically  

**Final Insight:**

- The highest ROI comes from structuring the business so the desired outcome happens by default

---

---

## 9.9.10 Final Executive Takeaway

- The business does not have a demand problem — it has a revenue capture problem  
- High-value behavior already exists but is not consistently maximized  
- The largest opportunities come from:
  - Increasing Average Order Value (AOV)
  - Monetizing off-peak demand
  - Removing peak-hour operational constraints  

- Sustainable growth is achieved by aligning:
  - Menu design  
  - Pricing structure  
  - Operational execution  

**Final Conclusion:**

- Revenue growth is not driven by more customers  
- It is driven by designing a system where higher-value orders occur by default

## 9.10 Executive Summary

- Revenue is primarily driven by high-value orders (30+ dollars), while mid-value orders (15–30 dollars) represent the largest opportunity for growth  
- Off-peak hours generate traffic but underperform in revenue due to low order values  
- Peak hours are demand-rich but constrained by operational throughput  

---

## Key Strategic Levers

- Increase Average Order Value (AOV) through add-ons and threshold-based nudging (e.g., cheese dip strategy)  
- Monetize off-peak demand using time-based, minimum-spend incentives  
- Improve peak-hour revenue by optimizing kitchen throughput and order flow  
- Align pricing and menu structure to guide customers into higher-value tiers  

---

## System Design Insight

- Revenue growth is not achieved by increasing traffic alone  
- The highest impact comes from designing a system where:

  - Customers naturally build higher-value orders  
  - Pricing reinforces spending thresholds  
  - Operations support demand efficiently  

---

## Final Recommendation

- Focus on AOV expansion before customer acquisition  
- Implement menu-driven and threshold-based strategies immediately  
- Use off-peak periods to capture additional revenue without increasing costs  
- Optimize peak-hour operations to prevent lost sales  

**Bottom Line:**

- The business already has demand  
- Growth comes from capturing more value per customer and structuring the system to make that outcome consistent
