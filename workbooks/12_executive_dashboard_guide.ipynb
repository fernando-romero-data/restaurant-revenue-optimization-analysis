# 12.0 Executive Dashboard Overview

The objective of this workbook is to translate the analytical findings established throughout the Buenavista Revenue Optimization System™ into executive-facing Tableau dashboards.

Unlike previous workbooks focused on exploration and strategic recommendations, Workbook 12 emphasizes communication, visualization, and decision support.

The dashboards developed in this phase are intended to provide ownership and leadership teams with a concise view of business performance while highlighting opportunities for operational improvement and revenue optimization.

The Tableau dashboards created through this workbook support:

- executive KPI reporting
- revenue trend monitoring
- revenue segmentation analysis
- operational performance evaluation
- resource prioritization
- leadership decision-making

The resulting dashboards will serve as the visual component of the Buenavista Revenue Optimization System™ and support Tableau Public publication, GitHub integration, LinkedIn distribution, and interview storytelling.

Primary dashboard deliverable:

**Buenavista Revenue Optimization Dashboard™**

## 12.2 Revenue Trend Analysis

### Business Question

How has revenue changed across reporting periods?

Understanding revenue trends allows leadership to evaluate overall business performance and identify periods of growth, stability, or decline.

Monitoring revenue over time provides context for strategic planning, staffing decisions, and future performance expectations.

### Recommended Visualization

Line Chart

### Required Fields

- month
- month_num
- order_total

### Tableau Configuration

- Columns: month
- Rows: SUM(order_total)
- Marks: Line
- Sort month using month_num to preserve chronological order

### Executive Interpretation

Revenue trend analysis provides leadership with a high-level view of business performance over time.

Periods of sustained growth may indicate successful operational execution, favorable market conditions, or effective revenue optimization strategies.

Periods of declining performance warrant further investigation into demand patterns, operational constraints, or changes in customer behavior.

Trend monitoring serves as an essential component of executive reporting and supports proactive decision-making.

### Strategic Importance

Revenue trends answer the question:

> "Is the business moving in the right direction?"

This visualization establishes the foundation for the remaining dashboard components by providing context for all subsequent analyses.

## 12.3 Revenue Segmentation Analysis

### Business Question

Which order segments contribute most to overall business performance?

Understanding how revenue is distributed across different order value segments enables leadership to prioritize strategies that maximize financial impact.

Revenue segmentation highlights whether growth opportunities lie in increasing customer traffic or increasing value captured from existing customers.

### Recommended Visualization

Horizontal Bar Chart

### Required Fields

- order_value_segment
- order_total

### Tableau Configuration

- Rows: order_value_segment
- Columns: SUM(order_total)
- Marks: Bar
- Sort descending by SUM(order_total)

Recommended segment order:

1. High Value (30+)
2. Mid Value (15–30)
3. Low Value (<15)

### Executive Interpretation

Revenue segmentation reveals that not all transactions contribute equally to business performance.

High Value orders generate a disproportionate share of total revenue despite representing a smaller percentage of overall transactions.

Mid Value orders represent the greatest opportunity for Average Order Value (AOV) expansion, as relatively small increases in spend can shift these orders into higher revenue tiers.

Low Value orders contribute the smallest share of revenue and should not be the primary focus of growth initiatives.

### Strategic Importance

Revenue segmentation answers the question:

> "Where should leadership focus first?"

The findings support a revenue optimization approach centered on increasing value captured from existing customers rather than relying exclusively on customer acquisition.

### Core Business Insight

The business does not have a demand problem.

The business has a revenue capture problem.

The largest opportunities for growth come from:

- expanding Average Order Value (AOV)
- converting Mid Value orders into High Value orders
- reinforcing behaviors associated with higher-value transactions

## 12.4 Operational Intelligence Analysis

### Business Question

How should operational resources be aligned to maximize revenue performance?

Understanding when revenue is generated allows leadership to prioritize staffing, operational efficiency, and revenue optimization initiatives.

Operational intelligence provides the foundation for identifying periods where the business should focus on throughput optimization versus demand activation.

### Recommended Visualization

Side-by-Side Bar Chart

### Required Fields

- operational_period
- order_total

### Tableau Configuration

- Columns: operational_period
- Rows: SUM(order_total)
- Marks: Bar
- Sort descending by SUM(order_total)

Recommended period order:

1. Peak
2. Off-Peak
3. Standard

### Executive Interpretation

Operational performance analysis demonstrates that revenue generation is not evenly distributed throughout the day.

Peak periods account for the majority of orders and revenue, indicating that these windows represent the greatest opportunity for throughput optimization.

Off-Peak periods contribute a meaningful share of overall revenue while operating below Peak demand levels, suggesting opportunities for increased utilization through targeted promotional initiatives.

Standard periods represent a minimal share of total business activity and require limited strategic focus.

### Strategic Importance

Operational intelligence answers the question:

> "Where should leadership prioritize operational resources?"

The findings support differentiated strategies based on operating conditions rather than applying a uniform approach throughout the day.

### Recommended Actions

**Peak Periods**

Focus on:

- maximizing throughput
- reducing operational bottlenecks
- ensuring appropriate staffing coverage
- streamlining high-frequency order execution

Goal:

Increase the number of orders fulfilled during periods of existing demand.

**Off-Peak Periods**

Focus on:

- increasing utilization
- activating incremental demand
- promoting meal-based ordering behavior
- encouraging revenue-generating add-ons

Goal:

Improve performance during underutilized operating windows without increasing fixed costs.

### Core Business Insight

Peak periods represent a throughput opportunity.

Off-Peak periods represent a utilization opportunity.

Different operational environments require different strategic responses.

## 12.5 Executive KPI Dashboard

### Business Question

What metrics should leadership monitor to evaluate overall business performance?

Executive KPI reporting provides a concise summary of business health and serves as the starting point for strategic discussions.

These metrics establish performance benchmarks and allow leadership teams to quickly assess whether the business is meeting operational and financial expectations.

### Recommended Visualization

KPI Cards

### Required Fields

- order_total
- order_id
- guest_count

### Tableau Configuration

**KPI Card 1 — Total Revenue**

- Calculation: SUM(order_total)
- Format: Currency

**KPI Card 2 — Total Orders**

- Calculation: COUNT(order_id)
- Format: Whole Number

**KPI Card 3 — Average Order Value (AOV)**

- Calculation: AVG(order_total)
- Format: Currency

**KPI Card 4 — Average Guests per Order**

- Calculation: AVG(guest_count)
- Format: One Decimal Place

### Executive Interpretation

Executive KPIs provide a high-level summary of business performance and establish the context for deeper analysis throughout the dashboard.

Total Revenue reflects the overall financial output of the business.

Total Orders measure transaction volume and overall demand.

Average Order Value (AOV) indicates how effectively the business captures value from each customer interaction.

Average Guests per Order provides insight into purchasing behavior and party size trends.

Together, these metrics provide leadership with a balanced view of performance across revenue generation, customer activity, and operational efficiency.

### Strategic Importance

Executive KPIs answer the question:

> "How is the business performing overall?"

These measures serve as the foundation of executive reporting and should be monitored consistently to identify emerging trends and evaluate the effectiveness of strategic initiatives.

### Recommended Dashboard Placement

Position KPI cards across the top of the Tableau dashboard to provide immediate context before presenting more detailed analyses.

Recommended order:

1. Total Revenue
2. Total Orders
3. Average Order Value (AOV)
4. Average Guests per Order

### Core Business Insight

No single metric tells the entire story.

Revenue performance should be evaluated through a combination of:

- financial outcomes
- transaction volume
- customer spending behavior
- operational indicators

This balanced approach supports more informed leadership decision-making.
